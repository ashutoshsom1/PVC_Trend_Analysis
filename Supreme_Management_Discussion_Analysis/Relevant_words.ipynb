```python
file=open(r'C:\Users\ashutosh.somvanshi\PVC_Trend_Analysis\Supreme_Management_Discussion_Analysis\Supreme_Text_files\tokens.txt', 'r')
text = file.read()
```

```python
import nltk
nltk.download('punkt')

from nltk.tokenize import word_tokenize

tokens = word_tokenize(text)
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))  # You can adjust the language as needed

filtered_words = [word for word in tokens if word.lower() not in stop_words]
```

```python
relevant_words = filtered_words
```
