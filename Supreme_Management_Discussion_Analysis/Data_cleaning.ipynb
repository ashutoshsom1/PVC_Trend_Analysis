````python
import os
import PyPDF4

def convert_pdf_to_text(pdf_path, txt_folder):
    with open(pdf_path, 'rb') as pdf_file:
        pdf_reader = PyPDF4.PdfFileReader(pdf_file)
        num_pages = pdf_reader.numPages
        txt_filename = f"pdf_.txt"
        txt_path = os.path.join(txt_folder, txt_filename)
        with open(txt_path, 'w', encoding='utf-8') as txt_file:
                
            for page_number in range(num_pages):
                page = pdf_reader.getPage(page_number)
                text = page.extractText()
                txt_file.write(text)
                txt_file.write('\n')  # Add a new line after each page
                # txt_filename = f"page_{page_number + 1}.txt"
                # txt_path = os.path.join(txt_folder, txt_filename)
                # txt_filename = f"page_{page_number + 1}.txt"
                
        
        # txt_filename = f"page_{page_number + 1}.txt"
        # txt_path = os.path.join(txt_folder, txt_filename)        txt_file.write(text)

    print('PDF converted to text successfully.')

# Provide the paths for your PDF file and the folder to save text files
pdf_file_path = r'C:\Users\ashutosh.somvanshi\PVC_Trend_Analysis\Supreme_Management_Discussion_Analysis\Supreme_pdf_files\annual-repoerts-2017-Dec.pdf'
txt_folder_path = r'C:\Users\ashutosh.somvanshi\PVC_Trend_Analysis\Supreme_Management_Discussion_Analysis\Supreme_text_files'+ '/'

# Create the folder if it doesn't exist
os.makedirs(txt_folder_path, exist_ok=True)

# Call the function to convert PDF to text and save in separate files
convert_pdf_to_text(pdf_file_path, txt_folder_path)
````

````python
import nltk

def split_string(text, max_tokens):
    tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
    tokens = tokenizer.tokenize(text)
    parts = []
    current_part = ""
    token_count = 0

    for token in tokens:
        if token_count + len(token.split()) > max_tokens:
            parts.append(current_part.strip())
            current_part = ""
            token_count = 0

        current_part += token + " "
        token_count += len(token.split())

    # Add the last part if any tokens remain
    if current_part.strip():
        parts.append(current_part.strip())

    return parts

# Example usage
cleaned_string = remove_special_characters(str(all_text_))
original_string = cleaned_string
parts = split_string(original_string, 3500)

# Print the parts
for i, part in enumerate(parts):
    part_variable_name = f"part_{i}"
    exec(f"{part_variable_name} = '{part}'")
    print(f"len__{len(part)}:{part_variable_name}: {part}")
````
